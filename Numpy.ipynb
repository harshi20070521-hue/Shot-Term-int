# Copies and Viewing

```python
v=np.arange(10)
w=v[::2]
print(v)
print(w)
```

```python
print(np.shares_memory(v,w))
```

```python
w[0]=20
print(v)
print(w)
```

```python
e=np.arange(10)
f=e[::2].copy()
print(e)
print(f)
```

```python
print(np.shares_memory(e,f))
```

```python
f[3]=25
print(e)
print(f)
```

### All Arithemtic operates Element wise

```python
r=np.ones(4)+1
print("a= ",a)
print("r= ",r)
```

```python
a-r
```

# Logical operators

```python
q=np.array([1,0,0,1],dtype="bool")
p=np.array([0,1,1,1],dtype="bool")
print(np.logical_or(q,p)) # logical or operation
```

```python
print(np.logical_and(p,q))#logical and operation
```

```python
print(np.logical_not(p)) #logical not operation
```
